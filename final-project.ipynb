```python
import matplotlib.pyplot as plt
```

```python
plt.bar([jt.name, jw.name, mg.name, do.name, cc.name], [jt.rush_yards, jw.rush_yards, mg.rush_yards, do.rush_yards, cc.rush_yards], width = 1)
plt.xlabel("Player Name")
plt.ylabel("Career Yards")
plt.xticks(rotation=30)
#includes total yards for career for some
# do rush_yards divided by games played
#James White worked more as receiver
```

```python
plt.bar([jt.name, jw.name, mg.name, do.name, cc.name], [(jt.rush_yards/jt.touches), (jw.rush_yards/jw.touches), (mg.rush_yards/mg.touches), (do.rush_yards/do.touches), (cc.rush_yards/cc.touches)], width = 1)
plt.xlabel("Player Name")
plt.ylabel("Career Yards/Touch")
plt.xticks(rotation=30)
```

```python
plt.scatter(
    x=[(jt.rush_yards/jt.touches), (jw.rush_yards/jw.touches), (mg.rush_yards/mg.touches), (do.rush_yards/do.touches), (cc.rush_yards/cc.touches)], 
    y = [jt.rush_yards, jw.rush_yards, mg.rush_yards, do.rush_yards, cc.rush_yards],
    c=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
plt.legend(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'], [jt.name, jw.name, mg.name, do.name, cc.name])
```

```python
name = [jt.name, jw.name, mg.name, do.name, cc.name]
yards = [jt.rush_yards, jw.rush_yards, mg.rush_yards, do.rush_yards, cc.rush_yards]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
ypt = [(jt.rush_yards/jt.touches), (jw.rush_yards/jw.touches), (mg.rush_yards/mg.touches), (do.rush_yards/do.touches), (cc.rush_yards/cc.touches)]
for i in range(0,5):
    plt.scatter(
    x= yards[i], 
    y = ypt[i],
    c=colors[i],
    label = name[i])
plt.legend()
```
